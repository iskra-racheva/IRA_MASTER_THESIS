```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# plotting things
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [22, 6]
plt.rcParams["figure.autolayout"] = True

liwi = 5.
```

```python
# sorting algorithms
quick = 'quick sort'
merge = 'merge sort'

# distributions
uni = 'Discrete Uniform'
uni_sort = 'Cont. Uniform Sorted'
binomial = 'Binomial'
uni_revsort = 'Cont. Uniform Reverse Sorted'
poisson = 'Poisson'
zipf = 'Zipf'
```

```python
dist_to_opt = {uni: quick, uni_sort: merge, \
               binomial: quick, uni_revsort: merge, \
               poisson: quick, zipf: quick}
dist_to_color = {uni: 'tab:blue', uni_sort: 'tab:orange', \
                 binomial: 'tab:gray', uni_revsort: 'tab:red', \
                 poisson: 'tab:cyan', zipf: 'tab:green'}
    
def plot_experiment(exp_num, changes, distributions, size_chunks):
    for idx in range(len(changes)-1):
        #print(changes[idx]*size_chunks, changes[idx+1]*size_chunks)
        plt.plot([changes[idx]*size_chunks, changes[idx+1]*size_chunks], [exp_num,exp_num], c=dist_to_color[distributions[idx]], marker="o", alpha=0.7, markersize="7", lw=liwi, label=distributions[idx])
    plt.ylabel("Experimental Setting",fontsize=26)
    plt.xlabel("Number of samples",fontsize=26)
    plt.yticks([1,2, 3,4,5], ['First Setting', 'Second Setting', 'Third Setting', "Fourth Setting", "Fifth Setting"],fontsize=26) 
    plt.ticklabel_format(axis="x",style='plain')
    plt.xticks(fontsize=26, rotation=45)
### 1st Experiment ####
size_chunks   = 1e+5
#num_dp        = 2e+8
#num_chunks    = num_dp / size_chunks
changes       = [0,1000,2000]
distributions = [uni,  uni_sort]

plt.figure()
plot_experiment(1, changes, distributions, size_chunks)
### 2nd Experiment ###
size_chunks   = 5e+5
changes       = [0,50,100]
distributions = [uni, uni_sort]
plot_experiment(2, changes, distributions, size_chunks)
### 3rd Experiment ###
size_chunks   = 5e+4
changes       = [0,1000, 2000,  3000, 4000]
distributions = [uni,  uni_sort, binomial, uni_revsort]
plot_experiment(3, changes, distributions, size_chunks)
### 4th Experiment ###
size_chunks   = 1e+5
changes       = [0,100, 150, 250, 300]
distributions = [uni_sort, poisson, uni_revsort, zipf]
plot_experiment(4, changes, distributions, size_chunks)
### 5th Experiment ###
size_chunks   = 1e+5
changes       = [0,50, 100, 150,  200, 250,300]
distributions = [uni_sort, poisson, uni_revsort, zipf, uni_revsort, uni]
plot_experiment(5, changes, distributions, size_chunks)   

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),fontsize=22, loc='center left', bbox_to_anchor=(1, 0.5)) 
```
