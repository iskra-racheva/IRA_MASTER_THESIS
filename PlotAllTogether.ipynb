```python
import matplotlib.pyplot as plt
import pandas as pd
lw = 3
```

```python
def plot_runtimes(df_quick_sort, df_merge_sort, df_heap_sort, labels, title):
    plt.figure(figsize=(10,6))
    #plt.scatter(labels,df_quick_sort, alpha = 0.5)
    plt.plot(labels, df_quick_sort, label="quick sort", alpha = 0.7, lw = lw, marker="o", markersize=10)
    #plt.scatter(labels,df_merge_sort, alpha = 0.5)
    plt.plot(labels, df_merge_sort, label="merge sort", alpha = 0.7, lw = lw, marker="o", markersize=10)
    #plt.scatter(labels,df_heap_sort, alpha = 0.5)
    plt.plot(labels, df_heap_sort, label="heap sort", alpha = 0.7, lw = lw, marker="o", markersize=10)
    plt.legend(loc='upper left',fontsize=24)
    #plt.title(title,fontsize=24)
    plt.ylabel("Runtime in seconds",fontsize=24)
    plt.xlabel("Number of samples",fontsize=24)
    plt.ticklabel_format(axis="x",style='plain')
    plt.xticks(fontsize=24,rotation=45)
    plt.yticks(fontsize=24)
    plt.show()
```

```python
labels = [10000,50000,100000,500000,1000000]
folder = "DATA/"
```

```python
def load_data(distribution, folder):
    mean_10000_quick_sort = (pd.read_csv(folder + distribution+ "_10000.csv")).quicksort.mean()
    mean_10000_merge_sort = (pd.read_csv(folder + distribution+ "_10000.csv")).mergesort.mean()
    mean_10000_heap_sort = (pd.read_csv(folder + distribution+ "_10000.csv")).heapsort.mean()

    mean_50000_quick_sort = (pd.read_csv(folder + distribution+ "_50000.csv")).quicksort.mean()
    mean_50000_merge_sort = (pd.read_csv(folder + distribution+ "_50000.csv")).mergesort.mean()
    mean_50000_heap_sort = (pd.read_csv(folder + distribution+ "_50000.csv")).heapsort.mean()

    mean_100000_quick_sort = (pd.read_csv(folder + distribution+ "_100000.csv")).quicksort.mean()
    mean_100000_merge_sort = (pd.read_csv(folder + distribution+ "_100000.csv")).mergesort.mean()
    mean_100000_heap_sort = (pd.read_csv(folder + distribution+ "_100000.csv")).heapsort.mean()

    mean_500000_quick_sort = (pd.read_csv(folder + distribution+ "_500000.csv")).quicksort.mean()
    mean_500000_merge_sort = (pd.read_csv(folder + distribution+ "_500000.csv")).mergesort.mean()
    mean_500000_heap_sort = (pd.read_csv(folder + distribution+ "_500000.csv")).heapsort.mean()

    mean_1000000_quick_sort = (pd.read_csv(folder + distribution+ "_1000000.csv")).quicksort.mean()
    mean_1000000_merge_sort = (pd.read_csv(folder + distribution+ "_1000000.csv")).mergesort.mean()
    mean_1000000_heap_sort = (pd.read_csv(folder + distribution+ "_1000000.csv")).heapsort.mean()
    
    quick_sort_summary = [mean_10000_quick_sort,mean_50000_quick_sort,mean_100000_quick_sort,mean_500000_quick_sort,mean_1000000_quick_sort]
    merge_sort_summary = [mean_10000_merge_sort,mean_50000_merge_sort,mean_100000_merge_sort,mean_500000_merge_sort,mean_1000000_merge_sort]
    heap_sort_summary = [mean_10000_heap_sort,mean_50000_heap_sort,mean_100000_heap_sort,mean_500000_heap_sort,mean_1000000_heap_sort]
    return quick_sort_summary,merge_sort_summary,heap_sort_summary
```

```python
# runtimes for discrete uniform distrbution
quick_sort_discrete_uniform,merge_sort_discrete_uniform,heap_sort_discrete_uniform = load_data("df_discrete_uniform", folder)
plot_runtimes(quick_sort_discrete_uniform, merge_sort_discrete_uniform, heap_sort_discrete_uniform, labels, "Sorting on discrete uniform distribution")
```
